```python
## Load in needed mod
import pandas as pd, numpy as np

## Set path to phenotype data
pheno_path = '../PHENOTYPE/RAW/F1-progeny-info.xlsx'

## Bring in phenotype data
data = pd.read_excel(pheno_path)
data.dropna(inplace=True)
data.reset_index(drop=True,inplace=True)
data.sort_values('Mutation rate (quantitative)',inplace=True)
data.index = [str(s) for s in data['F1 Progeny #'].values]

## Rescale phenotype data
scale = (10**7)
data['Rate'] = np.array(scale* data['Mutation rate (quantitative)'].values,dtype=float)
data['Uci'] = scale* data['Upper confidence interval'].values
data['Lci'] = scale* data['Lower confidence interval'].values

## Save gzipped dataframe
data.to_csv('../PHENOTYPE/F1-progeny-info-processed.csv.gz')

## View head
data.head().T
```
